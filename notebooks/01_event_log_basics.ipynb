# Event Log Basics

Load a CSV, convert it to a PM4Py event log, export to XES, and inspect process variants.

```python
import pandas as pd
import pm4py
```

## Load CSV Data
Read the HR onboarding CSV and parse the timestamp column.

```python
df = pd.read_csv("../../data/examples/HR_Onboarding.1.csv")

df["Timestamp"] = pd.to_datetime(df["Timestamp"])

df.head()
```

## Convert to Event Log
Use PM4Py to format the DataFrame with the correct column mappings and convert it to an event log object.

```python
formatted_df = pm4py.format_dataframe(
    df,
    case_id="Employee_ID",
    activity_key="Activity",
    timestamp_key="Timestamp"
)

event_log = pm4py.convert_to_event_log(formatted_df)
```

## Inspect Log Size
Check how many traces (cases) are in the event log.

```python
len(event_log)
```

## Export to XES
Write the event log to the standard XES format for interoperability with other process mining tools.

```python
pm4py.write_xes(
    event_log,
    "../../data/examples/HR_Onboarding.1.xes"
)
```

## Process Variants
Extract and display the unique activity sequences (variants) in the log.

```python
variants = pm4py.get_variants(event_log)

for i, (variant, traces) in enumerate(variants.items(), start=0):
    print(f"\nVariant #{i}")
    print(variant)
    print("Cases:", len(traces))
```
